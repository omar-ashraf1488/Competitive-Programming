```python

def main():
  n, iceCreamSum = map(int, input().split())
  kids = 0
  for i in range(n):
    signal, iceCream = map(str, input().split())
    if signal == "+":
      iceCreamSum += int(iceCream)
    else:
      if iceCreamSum < int(iceCream):
        kids += 1
      else:
        iceCreamSum -= int(iceCream)

  print(iceCreamSum, kids)
if __name__ == '__main__':
    main()
```
