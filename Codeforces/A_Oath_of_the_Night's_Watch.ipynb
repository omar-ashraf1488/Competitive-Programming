```python
def main():
  n = int(input())
  stewards = list(map(int, input().split()))[:n]

  stewards_min = min(stewards)
  stewards_max = max(stewards)
  count = 0
  for i in range(n):
    if stewards[i] > stewards_min and stewards[i] < stewards_max:
        count += 1
  print(count)
  
                    
if __name__ == '__main__':
    main()
```
