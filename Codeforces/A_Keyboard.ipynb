```python
def main():
  characters = "qwertyuiopasdfghjkl;zxcvbnm,./"

  direction = input()
  string = input()

  word = ""
  if direction =='R':
      for x in string:
          word += characters[characters.index(x) - 1]
  else:
      for x in string:
          word += characters[characters.index(x) + 1]
 
  print(word)
           
if __name__ == '__main__':
    main()
```
