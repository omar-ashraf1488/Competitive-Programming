```python

def main():
  n = int(input())
  listFriends = list(map(int, input().split()))[:n]
                     
  solution = []
  for i in range(n):
    solution.append(listFriends.index(i + 1) + 1)

  print(*solution)
        

if __name__ == '__main__':
    main()
```
