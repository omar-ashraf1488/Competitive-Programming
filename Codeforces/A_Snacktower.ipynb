```python
def main():
  n = int(input())
  snacksList = list(map(int, input().split()))[:n]
  sortSet = set() # To sort numbers in ascending order

  for i in snacksList:
    sortSet.add(i)
    while n in sortSet:
        print(n, end=' ')
        n -= 1
    print()
                    
if __name__ == '__main__':
    main()
```
