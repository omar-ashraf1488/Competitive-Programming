```python

def main():
    n, juicerSize, wasteSize = map(int, input().split())
    sumSizes=0
    count = 0
    orangeList = list(map(int, input().split()))

    for i in range(n):
        if orangeList[i] <= juicerSize:
            sumSizes += orangeList[i]
        if sumSizes > wasteSize:
            sumSizes = 0
            count += 1
    print(count)

if __name__ == '__main__':
    main()
```
