```python
def main():
  n = int(input())
  xCoordinate  = list(map(int, input().split()))[:n]

  for i in range(n):
    if i == 0:
      min1 = abs(xCoordinate[i] - xCoordinate[i + 1])
    elif i == n-1:
      min1 = abs(xCoordinate[i] - xCoordinate[i - 1])
    else:
      min1 = min(abs(xCoordinate[i] - xCoordinate[i + 1]) , abs(xCoordinate[i] - xCoordinate[i-1]))
    
    max1 = max(abs(xCoordinate[i] - xCoordinate[0]), abs(xCoordinate[i] - xCoordinate[n-1]))
    print(min1, max1)
                    
      
if __name__ == '__main__':
    main()
```
