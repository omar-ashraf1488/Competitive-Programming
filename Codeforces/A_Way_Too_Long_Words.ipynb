```python

def main():
  wordsList = []
  n = int(input())
  for i in range(n):
    wordsList.append(input())
  for i in range(n):
    if len(wordsList[i]) <= 10:
      print(wordsList[i])
    else:
      print(wordsList[i][0] + str(len(wordsList[i])-2) + wordsList[i][-1])

if __name__ == '__main__':
    main()
```
