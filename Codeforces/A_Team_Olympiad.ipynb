```python

def main():
  n = int(input())
  children = input().split()

  programming = []
  maths = []
  PE = []
  
  for i in range(len(children)):
    if children[i] == "1":
      programming.append(i+1)
    elif children[i] == "2":
      maths.append(i+1)
    else:
      PE.append(i+1)

  num = min(len(programming), len(maths), len(PE))
  print(num)
 
  for i in range(num):
    print(str(programming[i]) + " " + str(maths[i]) + " " + str(PE[i]))

if __name__ == '__main__':
    main()
```
